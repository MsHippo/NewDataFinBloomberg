### Add the GICS sector

```python
GICS = pd.ExcelFile("..\Sray\original dirty data\GICS sector by country (robusness check).xlsx")

'''CH'''
GICS_SECTOR = pd.read_excel(GICS, sheet_name="CH")

'''AU'''
# GICS_SECTOR = pd.read_excel(GICS, sheet_name="AU")

'''HK'''
# GICS_SECTOR = pd.read_excel(GICS, sheet_name="HK")

'''MY'''
# GICS_SECTOR = pd.read_excel(GICS, sheet_name="MK")

'''SG'''
# GICS_SECTOR = pd.read_excel(GICS, sheet_name="SG")

GICS_SECTOR_pd = pd.DataFrame(GICS_SECTOR)
GICS_SECTOR_pd.head()
```

```python
sec_no = pd.read_excel(GICS, sheet_name="Sector No")
sec_no_pd = pd.DataFrame(sec_no)

sec_no_pd = sec_no_pd.rename({'sector' : 'GICS Sector', 'sec_no': 'sec_no'.capitalize()}, axis=1)
sec_no_pd.head()
```

```python
GICS_SECTOR_pd = pd.merge(GICS_SECTOR_pd, sec_no_pd, on=['GICS Sector'])
GICS_SECTOR_pd = GICS_SECTOR_pd.rename({'Ticker': 'Tkr & Exch'}, axis=1)
GICS_SECTOR_pd = GICS_SECTOR_pd.drop(['GICS Sector'], axis=1)
GICS_SECTOR_pd = GICS_SECTOR_pd.sort_values(['Tkr & Exch'])

GICS_SECTOR_pd.head()
```

```python
firm_no_pd = pd.DataFrame(fin_esg_data_full['Tkr & Exch'])
firm_no_pd = firm_no_pd.drop_duplicates()
firm_no_pd = firm_no_pd.reset_index(drop=True)
firm_no_pd.index += 1
firm_no_pd.index.name = 'firm no'
firm_no_pd = firm_no_pd.reset_index()
firm_no_pd.tail()
```

```python
fin_esg_gics_data_full = pd.merge(fin_esg_data_full, firm_no_pd, on=['Tkr & Exch'], how='left')
fin_esg_gics_data_full = pd.merge(fin_esg_gics_data_full, GICS_SECTOR_pd, on=['Tkr & Exch'], how='left')
Sec_No_Column = fin_esg_gics_data_full.pop('Sec_no')
Firm_No_Column = fin_esg_gics_data_full.pop('firm no')
Name_No_Column = fin_esg_gics_data_full.pop('Name')
fin_esg_gics_data_full.insert(0, 'firm no', Firm_No_Column)
fin_esg_gics_data_full.insert(4, 'Name', Name_No_Column)
fin_esg_gics_data_full.insert(5, 'Sec_no', Sec_No_Column)
fin_esg_gics_data_full = fin_esg_gics_data_full.sort_values(['firm no', 'Year'])
fin_esg_gics_data_full.tail()
```
